```python
# Step 1: Define the Charging Station Data
charging_stations = [
    {
        "id": 1,
        "location": "rushikonda",
        "charger_type": "Fast Charger",
        "total_slots": 10,
        "available_slots": 3
    },
    {
        "id": 2,
        "location": "complex",
        "charger_type": "Standard Charger",
        "total_slots": 8,
        "available_slots": 5
    },
    {
        "id": 3,
        "location": "vizag",
        "charger_type": "Super Charger",
        "total_slots": 12,
        "available_slots": 2
    },
    {
        "id": 4,
        "location": "kadapa",
        "charger_type": "Super Charger",
        "total_slots": 12,
        "available_slots": 6
    },
    # Add more stations as needed
]


def search_charging_stations(location=None, charger_type=None, min_available_slots=1):
    results = charging_stations
    if location:
        results = [station for station in results if station['location'] == location]
    if charger_type:
        results = [station for station in results if station['charger_type'] == charger_type]
    results = [station for station in results if station['available_slots'] >= min_available_slots]
    
    return results


def book_charging_slot(station_id):
    for station in charging_stations:
        if station["id"] == station_id:
            if station["available_slots"] > 0:
                station["available_slots"] -= 1
                return f"Slot booked successfully at {station['location']} (Station ID: {station_id})"
            else:
                return "No available slots at this station."
    return "Station not found."

# Example Usage:
# 1. Search for stations with filters
search_results = search_charging_stations(location="Downtown", charger_type="Fast Charger")
print("Search Results:", search_results)

# 2. Book a slot at a station
booking_result = book_charging_slot(station_id=1)
print("Booking Result:", booking_result)

# 3. Check the updated station data
print("Updated Charging Stations Data:", charging_stations)
```
